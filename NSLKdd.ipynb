## we will use
* standard scalar for numerical values and
* LabelEncoder for categorical or object values

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder

numerical_columns = df.drop(columns = ['protocol_type','service','flag','label'], axis =1)
categorical_columns = df[['protocol_type','service','flag','label']]
numerical_columns.shape, categorical_columns.shape
```

```python
scalar = StandardScaler()
X1 = scalar.fit_transform(numerical_columns)
X1 = pd.DataFrame(X1)
X1.columns = numerical_columns.columns
X1
```

```python
encoder = LabelEncoder()
X2 = categorical_columns.apply(encoder.fit_transform)
print(X2)
```

```python
data = pd.concat([X1, X2.reset_index(drop=True)], axis=1)
data.columns = data.columns.astype(str)
```

## as we dont have classes we resort to unsupervised learning

```python
scalar2 = StandardScaler()
X = scalar2.fit_transform(data)
X = pd.DataFrame(X)
X.columns = data.columns
X
```

```python
X.info()
```

```python
from sklearn.ensemble import IsolationForest

forest = IsolationForest(n_estimators=100, contamination=0.01, random_state=42)
y_hat = forest.fit_predict(X)
```

```python
X['class'] = pd.DataFrame(y_hat)
```
